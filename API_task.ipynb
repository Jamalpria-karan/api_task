```python
import requests
import pandas as pd
import time
```

```python
df=pd.DataFrame()
```

```python
#getting the top rated movies data from TMDB API as directed by the youtube video.
#link for json viewer is given in below line--->
#https://api.themoviedb.org/3/movie/top_rated?api_key=05cf3c7ce5d988b68d223a58e59d578c&language=en-US&page=1
time.sleep(1)
for i in range(1,3):
  url="https://api.themoviedb.org/3/movie/top_rated?api_key="+api_key+"&language=en-US&page={}".format(i)
  response=requests.get(url).json()
  temp_df=pd.DataFrame(response['results'])[['id','original_language','popularity','release_date','vote_average','vote_count']]
  df=df.append(temp_df,ignore_index=True)
```
